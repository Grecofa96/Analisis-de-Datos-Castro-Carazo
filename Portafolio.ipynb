<h2 style='color:#56a5ff;text-align: left'>4.1 Manejo de valores faltantes</h2>

<p>Que se debe hacer para el manejo de valores faltante?</p>

<ol>
  <li>Identificar como se muestran los valores faltantes</li>
  <li>Contar la cantidad de valores faltantes en el set de datos</li>
  <li>Imputacion de valores numericos</li>
  <li>Imputacion de valores categoricos</li>
  <li>Eliminar filas</li>
</ol>

```python
dfAutos.replace("?", np.nan, inplace = True)
dfAutos.head(5)
```

<h2 style='color:#56a5ff;text-align: left'>4.2 Contar la cantidad de valores faltantes en el set de datos</h3>

```python
dfTemp = dfAutos.isnull()
dfTemp.head(5)
```

```python
for column in dfTemp.columns.values.tolist():
    print(column)
    print(dfTemp[column].value_counts())
```

<h3 style='color:#56a5ff;text-align: left'>4.3 Imputacion de valores numericos</h3>

```python
av_normalized = dfAutos["normalized"].astype("float").mean(axis=0)
av_bore = dfAutos["bore"].astype("float").mean(axis=0)
av_stroke = dfAutos["stroke"].astype("float").mean(axis=0)
av_horsepower = dfAutos["horsepower"].astype("float").mean(axis=0)
av_peak = dfAutos["peak"].astype("float").mean(axis=0)
```

```python
dfAutos["normalized"].replace(np.nan, av_normalized, inplace = True)
dfAutos["bore"].replace(np.nan, av_bore, inplace = True)
dfAutos["stroke"].replace(np.nan, av_stroke, inplace = True)
dfAutos["horsepower"].replace(np.nan, av_horsepower, inplace = True)
dfAutos["peak"].replace(np.nan, av_peak, inplace = True)
dfAutos.head(5)
```

<h3 style='color:#56a5ff;text-align: left'>4.4 Imputacion de valores categoricos</h3>

```python
dfAutos["doors"].value_counts().idxmax()
```

```python
dfAutos["doors"].replace(np.nan, "four", inplace = True)
dfAutos.head(5)
```

<h3 style='color:#56a5ff;text-align: left'>4.5 Eliminar filas con valores faltantes</h3>

```python
dfAutos.dropna(subset = ["price"], axis=0, inplace=True)
dfAutos.reset_index(drop=True, inplace=True)
```
