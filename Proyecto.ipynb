<p> 5 - Cual es el tipo de propiedades residenciales que se vende con mayor frecuencia, donde y con que costo promedio? </p>

```python
dfs = df_casas[df_casas['Residentialtype'] == 'Single Family']
pdfs = dfs['Saleamount'].mean()
pdfs = round(pdfs, 2)
town = dfs['Town'].value_counts().idxmax()

print(f"Costo promedio 'Single Family': ${pdfs:.2f}")
print(f"Lugar donde se venden 'Single Family': {town}")
```

```python
pdfs = dfs.groupby('Town')['Saleamount'].mean()
topciudades = pdfs.sort_values(ascending=False).head(5)
dftopciudades = dfs[dfs['Town'].isin(topciudades.index)]

plt.figure(figsize=(10, 6))
plt.barh(dftopciudades['Town'], dftopciudades['Saleamount'], color='green')
plt.title('Top 5 Propiedades con costo promedio mas alto')
plt.xlabel('Costo Promedio')
plt.ylabel('Ciudad')

plt.grid(True, axis='x')
plt.tight_layout()
plt.show()
```
